# cifar10
## Cifar10-Opacus

```python
with open("cifar10_opacus.txt", "r") as f:
    txt = f.readlines()
txt
```

```python
loss_no_dp = []
acc_no_dp = []

for i in range(6, 186, 2):
    loss = txt[i].split()[2]
    acc = txt[i].split()[4]
    loss_no_dp.append(float(loss))
    acc_no_dp.append(float(acc))
```

```python
loss_dp = []
acc_dp = []
eps_dp = []
delta_dp = []
for i in range(200, 470, 3):
    loss = txt[i].split()[2]
    acc = txt[i].split()[4]
    loss_dp.append(float(loss))
    acc_dp.append(float(acc))

for i in range(201, 471, 3):
    eps = txt[i].split()[3].split(",")[0]
    delta = txt[i].split()[6].split(")")[0]
    eps_dp.append(float(eps))
    delta_dp.append(float(delta))
delta_dp
```

```python
plt.figure()
plt.plot(loss_no_dp, label="Without DP")
plt.plot(loss_dp, label="With DP")
plt.legend()
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss with Cifar10-Opacus")
plt.show()
```

```python
plt.figure()
plt.plot(acc_no_dp, label="Without DP")
plt.plot(acc_dp, label="With DP")
plt.legend()
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Test Accuracy with Cifar10-Opacus")
plt.show()
```

# Runtime Analysis

```python
import seaborn as sns
```

```python
runtime = {
    "category": ["without DP", "with DP", "without DP", "with DP", "without DP", "with DP"],
    'x': ["Objax", "Objax", "Tf-privacy", "Tf-privacy", "Opacus", "Opacus"],
    "y": [9.4, 19.8, 0.9585579804011753, 7.31680337020329, 0.41, 0.96],
    
}
sns.set_theme(style="whitegrid")
plt.plot(figsize=(10,6))
ax = sns.barplot(x = 'x', y = 'y', hue = 'category', data = runtime,
            capsize = 0.5,             
            saturation = 8,             
            errcolor = 'gray', errwidth = 2,  
            ci = 68   
            )
ax.set(title="MNIST Average Epoch Runtime", xlabel='Framework', ylabel='Average Epoch Time (all epochs)')
plt.show()
```

```python
runtime = {
    "category": ["without DP", "with DP", "without DP", "with DP", "without DP", "with DP"],
    'x': ["Objax", "Objax", "Tf-privacy", "Tf-privacy", "Opacus", "Opacus"],
    "y": [58.929856051339044, 140.98809560934703, 2.3091790596644084, 12.824658343527052, 8.196942467159696, 5.239097314410739],
    
}
sns.set_theme(style="whitegrid")
plt.plot(figsize=(10,6))
ax = sns.barplot(x = 'x', y = 'y', hue = 'category', data = runtime,
            capsize = 0.5,             
            saturation = 8,             
            errcolor = 'gray', errwidth = 2,  
            ci = 68   
            )
ax.set(title="Cifar10 Average Epoch Runtime", xlabel='Framework', ylabel='Average Epoch Time (all epochs)')
plt.show()
```
